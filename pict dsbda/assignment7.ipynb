```python
import nltk

with open("sample.txt","r") as file:
    doc = file.read()
```

```python
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

```python
print(nltk.sent_tokenize(doc))
```

```python

def clean_doc(text):
    s=['.',',','"','(',')','%','?']
    t=""
    for letter in text:
        if letter not in s:
            t+=letter
    return t
        
doc = clean_doc(doc)
print(doc)
    
```

```python
token = nltk.word_tokenize(doc)
print(token)
```

```python
pos_tokens = nltk.pos_tag(token)
pos_tokens
```

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))

# Filter out the stop words
filtered_tokens=[]
for word in token:
    if word.lower() not in stop_words:
        filtered_tokens.append(word)
token = filtered_tokens
print(filtered_tokens)
```

```python
from nltk.stem import PorterStemmer,WordNetLemmatizer
nltk.download('wordnet')

stem1=PorterStemmer()
lemm1=WordNetLemmatizer()

token_stem =[]
for word in token:
    token_stem.append(stem1.stem(word))
print(token_stem)
token_lemtize =[]
for word in token:
    token_lemtize.append(lemm1.lemmatize(word))
print(token_lemtize)

```

```python
def calculateTF(token):
    term_freq = {}
    for word in token:
        if word not in term_freq:
            term_freq[word] = token.count(word) / len(token)

    return term_freq

calculateTF(token)
```
