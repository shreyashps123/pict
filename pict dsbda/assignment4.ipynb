```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error

```

The R2 score is a number between 0 and 1. If it's closer to 1, it means the line your friend drew is very close to the actual data points, so it's a good fit. If it's closer to 0, it means the line is not a good fit.

The lower the MSE, the closer the data points are to the line, which means the line is a better fit.
Formula: MSE is calculated using the following formula: 𝑀 𝑆 𝐸 = 1 𝑛 ∑ 𝑖 = 1 𝑛 ( 𝑦 𝑖 − 𝑦 ^ 𝑖 ) 2 MSE= n 1 ​ ∑ i=1 n ​ (y i ​ − y ^ ​ i ​ ) 2 where 𝑦 𝑖 y i ​ is the actual value of the data point, 𝑦 ^ 𝑖 y ^ ​ i ​ is the predicted value of the data point by the linear regression model, and 𝑛 n is the number of data points

```python
x=df['RM'].values.reshape(-1,1)
y=df['MEDV'].values.reshape(-1,1)
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

# # Create and train the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# # Evaluate the model
y_pred_train = model.predict(X_train) 
y_pred = model.predict(X_test)
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
print(f'Coefficient: {model.coef_}')
print(f'Intercept: {model.intercept_}')
print(r2)
print(mse)
```
