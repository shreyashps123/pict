```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,classification_report

from sklearn.naive_bayes import GaussianNB
```

```python
x = df[["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]]
y = df[["Species"]].values.reshape(-1,1)

scaler = StandardScaler()
x =scaler.fit_transform(x)

```

```python
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state=42)

```

```python
model = GaussianNB()
model.fit(X_train,y_train)
```

```python
pred_y  = model.predict(X_test)
```

```python
cf=confusion_matrix(y_test,pred_y)

print(cf)
```

```python
tp_1 = cf[0][0]
fp_1 = cf[1][0]+cf[2][0]
fn_1 = cf[0][1]+cf[0][2]
tn_1 =cf.sum()-tp_1-fp_1-fn_1
```

```python
tp_2 = cf[1][1]
fp_2 = cf[0][1]+cf[2][1]
fn_2 = cf[1][0]+cf[1][2]
tn_2 = cf.sum()- tp_2 - fp_2 - fn_2
```

```python
print(f"Accuracy score is {accuracy_score(y_test, pred_y)}")
```

```python
print(f"Error rate is {1 - accuracy_score(y_test, pred_y)}")
```

```python
print(f"Precision score is {precision_score(y_test, pred_y, average='macro')}")
```

```python
print(f"Recall score is {recall_score(y_test, pred_y, average='macro')}")
```

```python
print(classification_report(y_test, pred_y))
```
